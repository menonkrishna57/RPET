```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
uniform_data=stats.uniform.rvs(size=100000,loc=0,scale=10)
```

```python
pd.DataFrame(uniform_data).plot(kind="density",figsize=(9,9),xlim=(-1,11))
```

```python
stats.uniform.cdf(
    x=2.5,
    loc=0,
    scale=10
)
```

```python
for x in range(-1,12,3):
    print("Density at x value " +str(x))
    print(stats.uniform.pdf(x,loc=0,scale=10))
```

```python
normal_data = stats.norm.rvs(size=100000,loc=-20,scale=1)
pd.DataFrame(normal_data).plot(kind="density",figsize=(9,9))
```
